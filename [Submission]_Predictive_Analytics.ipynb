# **Data Preparation**

**Encoding Fitur Kategori**

```python
# Melakukan one-hot encoding atau dummy encoding pada data kategorik
data = pd.concat([data, pd.get_dummies(data[categorical_features[0]], prefix='WeekStatus')],axis=1)
data = pd.concat([data, pd.get_dummies(data[categorical_features[1]], prefix='Day_of_week')],axis=1)
data = pd.concat([data, pd.get_dummies(data[categorical_features[2]], prefix='Load_Type')],axis=1)
data.drop(['WeekStatus','Day_of_week','Load_Type'], axis=1, inplace=True)
data = data.replace({True: 1, False: 0})
data.head()
```

Melakukan encoding fitur kategorik menggunakan one-hot encoding agar tidak terdapat asumsi ordinal pada kategori.

**Reduksi Dimensi dengan PCA**

```python
from sklearn.decomposition import PCA

# Reduksi dimensi menggunakan PCA pada fitur Lagging_Current_Reactive_Power_kVarh dan CO2(tCO2)
pca = PCA(n_components=1, random_state=123)
pca.fit(data[['Lagging_Current_Reactive_Power_kVarh','CO2(tCO2)']])
data['dimension_1'] = pca.transform(data.loc[:, ('Lagging_Current_Reactive_Power_kVarh','CO2(tCO2)')]).flatten()
data.drop(['Lagging_Current_Reactive_Power_kVarh','CO2(tCO2)'], axis=1, inplace=True)
```

Menghilangkan korelasi antar variabel fitur dengan reduksi dimensi menggunakan PCA pada variabel Lagging_Current_Reactive_Power_kVarh dan CO2(tCO2) kemudian mengambil satu principal component sebagai representasi dari dua fitur tersebut.

```python
# Reduksi dimensi menggunakan PCA pada fitur Leading_Current_Reactive_Power_kVarh dan Leading_Current_Power_Factor
pca = PCA(n_components=1, random_state=123)
pca.fit(data[['Leading_Current_Reactive_Power_kVarh','Leading_Current_Power_Factor']])
data['dimension_2'] = pca.transform(data.loc[:, ('Leading_Current_Reactive_Power_kVarh','Leading_Current_Power_Factor')]).flatten()
data.drop(['Leading_Current_Reactive_Power_kVarh','Leading_Current_Power_Factor'], axis=1, inplace=True)
```

Menghilangkan korelasi antar variabel fitur dengan reduksi dimensi menggunakan PCA pada Leading_Current_Reactive_Power_kVarh dan Leading_Current_Power_Factor kemudian mengambil satu principal component sebagai representasi dari dua fitur tersebut.

**Train-Test-Split**

```python
from sklearn.model_selection import train_test_split

# Split data menjadi train 90% dan test 10%
X = data.drop(["Usage_kWh"],axis =1)
y = data["Usage_kWh"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 123)
```

Melakukan split pada dataset menjadi data latih untuk melatih model dan data uji untuk menguji performa model sebesar 90:10. Data latih hanya 10% karena dataset cukup besar sehingga data latih sudah mencakup banyak variasi.

```python
print(f'Total # of sample in whole dataset: {len(X)}')
print(f'Total # of sample in train dataset: {len(X_train)}')
print(f'Total # of sample in test dataset: {len(X_test)}')
```

**Normalisasi**

```python
from sklearn.preprocessing import MinMaxScaler

numerical_features = ['Lagging_Current_Power_Factor', 'NSM', 'dimension_1', 'dimension_2']
scaler = MinMaxScaler()

# Fit MinMaxScaler pada data train
scaler.fit(X_train[numerical_features])
X_train[numerical_features] = scaler.transform(X_train.loc[:, numerical_features])
X_train[numerical_features].head()
```

Melakukan normalisasi skala pada varibel numerik dan komponen hasil reduksi dimendsi menjadi nilai antara 0 dan 1.

# **Model Development**

**KNN**

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor

# Membangun model KNN
knn = KNeighborsRegressor(
    n_neighbors=5,
    weights='uniform',
    metric='minkowski',
    p=2
)
knn.fit(X_train, y_train)
```

Membangun dan melatih model KNN Regressor dengan fit model ke data latih.

**Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor

# Membangun model Random Forest Regressor
rf = RandomForestRegressor(
    n_estimators=100,
    criterion='squared_error',
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features=1.0,
    bootstrap=True,
    random_state=42,
)
rf.fit(X_train, y_train)
```

Membangun dan melatih model Random Forest Regressor dengan fit model ke data latih.

**Gradient Boosting**

```python
from sklearn.ensemble import GradientBoostingRegressor

# Membangun model Gradient Boosting Regressor
gb = GradientBoostingRegressor(
    loss='squared_error',
    learning_rate=0.1,
    n_estimators=100,
    min_samples_split=2,
    min_samples_leaf=1,
    max_depth=3,
    max_features=None,
    random_state=42
)
gb.fit(X_train, y_train)
```

Membangun dan melatih model Gradient Boosting Regressor dengan fit model ke data latih.

**Evaluasi**

```python
# Fit MinMaxScaler pada data test
X_test.loc[:, numerical_features] = scaler.transform(X_test[numerical_features])
```

Melakukan fit preprocessing berupa normalisasi ke data latih.

```python
# Buat variabel mse dan R² yang isinya adalah dataframe nilai mse dan R² data train dan test pada masing-masing algoritma
mse = pd.DataFrame(columns=['train', 'test'], index=['KNN', 'RF', 'GB'])
r2 = pd.DataFrame(columns=['train', 'test'], index=['KNN', 'RF', 'GB'])

# Buat dictionary untuk setiap algoritma yang digunakan
model_dict = {'KNN':knn, 'RF':rf, 'GB':gb}

# Hitung metrik MSE dan R² untuk setiap model
for name, model in model_dict.items():
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    mse.loc[name, 'train'] = mean_squared_error(y_train, y_pred_train) / 1e3
    mse.loc[name, 'test'] = mean_squared_error(y_test, y_pred_test) / 1e3

    r2.loc[name, 'train'] = r2_score(y_train, y_pred_train)
    r2.loc[name, 'test'] = r2_score(y_test, y_pred_test)
```

Menyimpan metrik MSE dan R Squared hasil evaluasi ketiga model ke dalam variabel baru mse dan r2 dan dibagi 1000 untuk menyederhanakan nilai.

```python
# Tampilan bar plot pada nilai MSE masing-masing model
fig, ax = plt.subplots()
mse.sort_values(by='test', ascending=False).plot(kind='barh', ax=ax, zorder=3)
ax.grid(zorder=0)

mse
```

Terlihat bahwa nilai metrik mse pada model Random Forest paling kecil dibandingkan model KNN dan Gradient Boosting. Hal ini menunjukkan bahwa hasil prediksi dari Random Forest paling dekat dengan data asli sehingga Random Forest adalah yang paling baik diantara KNN dan Gradient Boosting.

```python
# Tampilan bar plot pada nilai R² masing-masing model
fig, ax = plt.subplots()
r2.sort_values(by='test', ascending=False).plot(kind='barh', ax=ax, zorder=3)
ax.grid(zorder=0)

r2
```

Terlihat bahwa nilai metrik R squared pada model Random Forest paling besar dibandingkan model KNN dan Gradient Boosting. Hal ini menunjukkan bahwa Random Forest dapat lebih efektif dalam memahami variansi data dan menghasilkan prediksi yang lebih akurat pada data.
